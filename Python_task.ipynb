# **Load Dataset ➡️ malware-detection**

```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Download the dataset (returns local path)
path = kagglehub.dataset_download("nsaravana/malware-detection")

print("Dataset downloaded to:", path)

# Now load the CSV manually with Pandas
df = pd.read_csv(path + "/Malware dataset.csv")

print("Dataset shape:", df.shape)
print(df.head())
```

# **write python code to print the shape of my dataset**

```python
print("Shape:", df.shape)                           # (1) Shape of dataset
```

# **give me code to list all column names in the dataset**

```python
print("Columns:", df.columns.tolist())             # (2) Column names
```

# **python code to display datatypes of every column**

```python
print("Data Types:\n", df.dtypes)                  # (3) Data types
```

# **give code to check missing values in each column**

```python
print("Missing values:\n", df.isnull().sum())      # (4) Missing values
```

# **code to count duplicate rows in dataframe**

```python
print("Duplicates:", df.duplicated().sum())        # (5) Duplicate rows
```

# **print summary statistics of all numeric columns**

```python
print("Summary Stats:\n", df.describe())           # (6) Descriptive stats
```

# **code to show distribution of classes in classification**

```python
print("Class distribution:\n", df['classification'].value_counts())  # (7)
```

# **python code to show frequency count of malware vs benign**

```python
df['classification'].value_counts(normalize=True).plot.pie(autopct='%1.1f%%')  # (8)
plt.title("Malware vs Benign Distribution")
plt.show()
```

# **draw a pie chart to show percentage of malware vs benign**

```python
sns.countplot(x='classification', data=df)         # (9) Countplot of target
plt.title("Count of Malware vs Benign")
plt.show()
```

# **make a countplot for classification column (malware vs benign)**

```python
numeric_cols = df.select_dtypes(include=np.number).columns

df[numeric_cols].hist(figsize=(15, 10), bins=30)   # (10) Histograms
plt.show()
```

# **plot boxplot of utime column grouped by classification**

```python
sns.boxplot(x='classification', y='utime', data=df)  # (11) Boxplot example
plt.title("utime distribution by class")
plt.show()
```

# **write code to create correlation heatmap for numeric columns**

```python
corr = df[numeric_cols].corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")                 # (12) Correlation heatmap
plt.title("Feature Correlation")
plt.show()
```

# **find top 5 features most correlated with utime**

```python
print("Top correlated with utime:\n", corr['utime'].sort_values(ascending=False).head(5))  # (13)
```

# **code to calculate skewness and kurtosis for numeric features**

```python
skewness = df[numeric_cols].skew()                 # (14) Skewness
```

```python
kurtosis = df[numeric_cols].kurt()                 # (15) Kurtosis
print("Skewness:\n", skewness)
print("Kurtosis:\n", kurtosis)
```

# **how to list all categorical columns in dataset using python**

```python
cat_cols = df.select_dtypes(exclude=np.number).columns
print("Categorical columns:\n", cat_cols)          # (16) List categorical features
```

# **give me code to count unique values for each categorical column**

```python
for col in cat_cols:
    print(col, ":", df[col].nunique(), "unique values")   # (17)
```

# **make countplot of state column with classification as hue**

```python
sns.countplot(x='state', hue='classification', data=df)   # (18) Example categorical relation
plt.title("State vs Classification")
plt.show()
```

# **show code to calculate mean of numeric columns grouped by classification**

```python
print("Mean features by class:\n", df.groupby('classification')[numeric_cols].mean())   # (19)
```

# **find variance of all numeric features and display top 5 with highest variance**

```python
variances = df[numeric_cols].var().sort_values(ascending=False)   # (20)
print("Top 5 features by variance:\n", variances.head())
```
